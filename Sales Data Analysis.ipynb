<h3> EXPLORATORY DATA ANALYSIS

**Q1. What is the overall sales trend?**

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
# Getting month year from order date
df['month_year'] = df['order_date'].apply(lambda x:x.strftime('%Y-%m'))
```

```python
df['month_year']
```

```python
# Grouping month_year
df_trend = df.groupby('month_year').sum(numeric_only=True)['sales'].reset_index()
```

```python
# Linechart of the sales trend + Setting the figure size
plt.figure(figsize=(16,9))
plt.plot(df_trend['month_year'], df_trend['sales'])
plt.xticks(rotation = 90, size = 8)
plt.show()
```

**INSIGHT :** 

**Overall the sales number are increasing but in the initial 5 months of every year the sales number are low compared to the later half of the year.**

**Q2. Which are the Top 10 products by sales?**

```python
# Grouping products by sales
product_sales = pd.DataFrame(df.groupby('product_name').sum(numeric_only=True)['sales'])
# Sorting the dataframe in descending order
product_sales = product_sales.sort_values(by='sales', ascending=False)
# Top 10 products by sales
product_sales[:10]
```

**Q3. Which are the Most Selling Products?**

```python
# Grouping products by Quantity
product_quantity = pd.DataFrame(df.groupby('product_name').sum(numeric_only=True)['quantity'])
# Sorting the dataframe in descending order
product_quantity = product_quantity.sort_values('quantity',ascending=False)
# Most selling products
product_quantity[:10]
```

**Q4. Which is the most preferred Ship Mode?**

```python
# Setting the figure size
plt.figure(figsize=(10, 6))
# Using Countplot to plot and find out the most used ship mode
sb.countplot(df,x='ship_mode')

plt.show()
```

**Q5. Which are the Most Profitable Category and Sub-Category?**

```python
# Grouping products by Category and Sub-Category
product_category = pd.DataFrame(df.groupby(['category','sub_category']).sum(numeric_only=True)['profit'])
# Sorting the values
product_category.sort_values(['category','profit'],ascending=False)
```
